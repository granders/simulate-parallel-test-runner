```python
def plot_wallclock(on_demand, partitioned):
    fig, ax = plt.subplots()
    
    r1 = on_demand
    x1 = np.array(r1['nodes'])
    y1 = np.array(r1['wallclock'])

    r2 = partitioned
    x2 = np.array(r2['nodes'])
    y2 = np.array(r2['wallclock'])

    ax.plot(x1, y1)
    ax.plot(x2, y2)
    ax.plot(x2, np.array([120]* len(x2)))
    ax.plot(x2, np.array([180]* len(x2)))

    ax.grid(True)

# fig, ax = plt.subplots()

# ax.grid(True)
```

```python
def cluster_bringup_minutes(num_nodes):
    return 2.5 * math.ceil(num_nodes / 5.0)
```

```python
def plot_wallclock_with_cluster_bringup(on_demand, partitioned):
    fig, ax = plt.subplots()

    r1 = on_demand
    x1 = np.array(r1['nodes'])
    adjusted_wallclock = [r1['wallclock'][i] + cluster_bringup_minutes(r1['nodes'][i]) for i in range(len(r1['nodes']))]
    y1 = np.array(adjusted_wallclock)

    r2 = partitioned
    x2 = np.array(r2['nodes'])
    adjusted_wallclock = [r2['wallclock'][i] + cluster_bringup_minutes(r2['nodes'][i]) for i in range(len(r2['nodes']))]
    y2 = np.array(adjusted_wallclock)

    ax.plot(x1, y1)
    ax.plot(x2, y2)
    ax.plot(x2, np.array([120]* len(x2)))
    ax.plot(x2, np.array([180]* len(x2)))
    ax.grid(True)    
```

```python
for i in range(len(on_demand_results)):
    plot_wallclock_with_cluster_bringup(on_demand_results[i], partitioned_results[i])
```
